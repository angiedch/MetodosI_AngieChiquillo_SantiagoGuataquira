```python
#Punto 2 parte 1 
def piezas(n,a):
    p=np.linspace(0,1,10000)
    prob_aceptacion=[]
    for i in p:
        y=np.random.binomial(n,i,10000)
        aceptados = np.sum(y <= a)
        prob_aceptacion.append(aceptados/10000)
                
    
    
    
    
    return prob_aceptacion

p=np.linspace(0,1,10000)
modelo1= piezas(5,1)
modelo2= piezas(25,5)
fig,axs=plt.subplots(1,2,figsize=(20,5))
axs[0].plot(p,modelo1,color="#d61ca8")
axs[0].set_ylabel('Probabilidad de aceptación')
axs[0].set_xlabel('Piezas defectuozas p')
axs[0].set_title("Modelo (5,1)")

axs[1].plot(p,modelo2,color="#d61ca8")
axs[1].set_ylabel('Probabilidad de aceptación')
axs[1].set_xlabel('Piezas defectuozas p')
axs[1].set_title("Modelo (25,5)")
piezas(5,1) 

"Si fueras un venderdor lo ideal sería usar el primer modelo ya que este asegura una producción uniforme de piezas defectuosas, por otro" 
"lado  si fueras comprador prefieres el segundo modelo ya que despues de un p 0.3 las posibilidades de una pieza defectuosa son muy bajas "
```
