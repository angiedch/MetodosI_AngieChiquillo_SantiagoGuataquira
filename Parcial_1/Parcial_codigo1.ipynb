```python
import numpy as np
import matplotlib.pyplot as plt

def funcion(x):
    return (np.e**-x)*-x

x= np.linspace(-1,1,50)
y= funcion(x)
print(x)
plt.plot(x,y)
plt.axhline(y = 0,color='b')

```

```python
def metodo_muller(f, x, itmax=100, precision=1e-10):
    iteracion= 0
    error= 1
    aprox= float()
    
    while error > precision and iteracion < itmax:
        y0= f(x[0])
        y1= f(x[1])
        y2= f(x[2])
        
        f1= (y1-y0)/(x[1]-x[0])
        f2= (y2-y1)/(x[2]-x[1])

        f2= (f2-f1)/(x[2]-x[0])
        
        a= f2
        b= f1 + ((x[0]+x[1])*f2)
        c= y0 - ( x[0] * f1) + (x[0]*x[1]*f2)
        
        d= (b**2 )- (4*a*c)

        aprox_1= (-2*c)/(b+np.sqrt(d))
        aprox_2= (-2*c)/(b-np.sqrt(d))
        
        if b < 0:
            aprox= aprox_2
            
        elif b >= 0:
            aprox= aprox_1
        
        iteracion+=1
        error= np.abs(f(aprox))
        
        
    return aprox
        
print(metodo_muller(funcion, x, 1.))      

            
```
