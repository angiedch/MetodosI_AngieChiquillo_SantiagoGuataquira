```python
import numpy as np
import matplotlib.pyplot as mp 
import pandas as pd 
import sympy as sp
datos=pd.read_csv("Parabolico.csv",sep=",")
X=np.array(datos.X)
Y=np.array(datos.Y)

def Lagrange(x,X,i):
    
    L = 1
    
    for j in range(X.shape[0]):
        if i != j:
            L *= (x - X[j])/(X[i]-X[j])
            
    return L
def polinomio(x,X,Y):
    
    Pol = 0
    
    for i in range(X.shape[0]):
        Pol += Lagrange(x,X,i)*Y[i]
        
    return Pol




x=np.linspace(0,6,50)
y1= polinomio(x, X, Y)
mp.plot(x,y1)
mp.scatter(X, Y, color='g')


_x=sp.Symbol('x',real=True)
f =polinomio(_x, X, Y)
f= sp.simplify(f)
f

def funcion(x):
    return x*(0.363970234266202 - 0.0554912422401579* x)
z=funcion(x)
def DefCentral(f, x, h):
    return (f(x+h) - f(x-h))/(2*h)

v= DefCentral(funcion,x,0.1)
mp.plot(x,v)
print(v)
print(y1)
#en t=0.4007954931819738 x=1.4
magnitudv0=(1.4/(0.364*0.400))
angulo=np.arcsin((9.81*0.400)/(2*magnitudv0))
print(magnitudv0)




```
